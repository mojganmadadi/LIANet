```python
import json
import numpy as np
import matplotlib.pyplot
import os
import glob
import json
```

```python
topdir = "/home/user/results_local/finetuning_results"
```

```python
def read_file_and_take_best(topdir, task, model, minmax):
    path = os.path.join(topdir, task, model, "*", "best_metrics.json")
    files = glob.glob(path, recursive=True)

    # if len(files) != 3:
    #     print(len(files))
    #     return "Not Three Files Found"
    
    metrics_list = []

    for file in files:
        with open(file, "r") as f:
            data = json.load(f)
            metrics_list.append(data["best_values"])

    # Collect all keys
    keys = metrics_list[0].keys()
    results = {}

    for key in keys:
        values = [m[key] for m in metrics_list if key in m]

        best = 99
        if minmax == "max":
            best = float(np.max(values))
        else:
            best = float(np.min(values))
        
        results[key] = {
            "best": best,
        }

    return results


def print_classification(metric_dict):
    if metric_dict == "Not Three Files Found":
        return "Not Three Files Found"
    # metrics to include
    metrics = ["jaccard_macro", "accuracy_macro", "f1_macro"]

    values = []
    for m in metrics:
        mean = metric_dict[m]["best"]
        values.append(f"{mean:.2f}")

    return " & ".join(values)


def print_regression(metric_dict):
    if metric_dict == "Not Three Files Found":
        return "Not Three Files Found"
    metrics = ["mae", "mse"]

    values = []
    for m in metrics:
        mean = metric_dict[m]["best"]
        if m == "mse": # make it RSMSE
            #mean = np.sqrt(mean)
            pass
        values.append(f"{mean:.3f}")
        
    return " & ".join(values)
```
